```python
import torch
from torch.utils.data import Dataset
import pandas as pd
import librosa
import cv2
import numpy as np
from transformers import AutoTokenizer

class MultiModalDataset(Dataset):
    def __init__(self, excel_path, label_map):
        self.df = pd.read_excel(excel_path)
        self.tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")
        self.label_map = label_map

    def extract_audio(self, path):
        try:
            y, sr = librosa.load(path, sr=16000)
            if len(y) == 0:
                raise Exception("empty audio")
            # fix length to 1 sec (16000)
            if len(y) < 16000:
                y = np.pad(y, (0, 16000 - len(y)))
            else:
                y = y[:16000]
            return torch.tensor(y, dtype=torch.float32)
        except:
            return torch.zeros(16000, dtype=torch.float32)

    def extract_video(self, path):
        try:
            cap = cv2.VideoCapture(path)
            frames = []

            for _ in range(10):
                ret, frame = cap.read()
                if not ret:
                    break
                frame = cv2.resize(frame, (224, 224))
                frames.append(frame)

            cap.release()

            if len(frames) == 0:
                raise Exception("no frames")

            frames = np.array(frames)

            # pad to 10 frames
            if frames.shape[0] < 10:
                pad = np.zeros((10 - frames.shape[0], 224, 224, 3))
                frames = np.concatenate([frames, pad], axis=0)

            return torch.tensor(frames, dtype=torch.float32)

        except:
            return torch.zeros((10, 224, 224, 3), dtype=torch.float32)

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        text = str(row["Hindi text"])
        label = self.label_map[str(row["intent"]).strip()]

        enc = self.tokenizer(
            text,
            padding="max_length",
            truncation=True,
            max_length=128,
            return_tensors="pt"
        )

        audio = self.extract_audio(row["audio_path"])
        video = self.extract_video(row["video_path"])

        return {
            "input_ids": enc["input_ids"].squeeze(0),
            "attention_mask": enc["attention_mask"].squeeze(0),
            "audio": audio,
            "video": video,
            "label": torch.tensor(label, dtype=torch.long)
        }
```

```python
import torch.nn as nn
from transformers import AutoModel
import torchvision.models as models
import torch

class MultiModalModel(nn.Module):
    def __init__(self, num_labels):
        super().__init__()

        # TEXT
        self.text_model = AutoModel.from_pretrained("bert-base-multilingual-cased")

        # AUDIO
        self.audio_fc = nn.Linear(16000, 128)

        # VIDEO
        self.video_model = models.resnet18(pretrained=True)
        self.video_model.fc = nn.Linear(512, 128)

        # FUSION
        self.fc = nn.Linear(768 + 128 + 128, num_labels)

    def forward(self, input_ids, attention_mask, audio, video):

        # ---- TEXT ----
        text_out = self.text_model(input_ids=input_ids, attention_mask=attention_mask)
        text_feat = text_out.pooler_output  # (B, 768)

        # ---- AUDIO ----
        audio = audio.float()
        audio_feat = self.audio_fc(audio)   # (B, 128)

        # ---- VIDEO SAFE HANDLING ----
        if video.dim() == 4:
            video = video.unsqueeze(0)

        if video.dim() != 5:
            batch_size = input_ids.size(0)
            video = torch.zeros((batch_size, 10, 224, 224, 3)).to(input_ids.device)

        video = video.float()
        video = video.permute(0, 1, 4, 2, 3)  # (B, T, C, H, W)

        video_feats = []
        for i in range(video.shape[1]):
            frame_feat = self.video_model(video[:, i])
            video_feats.append(frame_feat)

        video_feat = torch.stack(video_feats, dim=1).mean(dim=1)  # (B, 128)

        # ---- FUSION ----
        combined = torch.cat([text_feat, audio_feat, video_feat], dim=1)
        return self.fc(combined)
```

```python
from torch.utils.data import DataLoader
import torch
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# PATH
EXCEL_PATH = "/content/drive/MyDrive/EDUINTENT/manifest.xlsx"

# Load labels
df = pd.read_excel(EXCEL_PATH)
labels = sorted(df["intent"].astype(str).str.strip().unique())
label_map = {l:i for i,l in enumerate(labels)}

dataset = MultiModalDataset(EXCEL_PATH, label_map)
loader = DataLoader(dataset, batch_size=4, shuffle=True)

model = MultiModalModel(len(label_map))
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)
loss_fn = torch.nn.CrossEntropyLoss()

# ---- TRAIN ----
for epoch in range(3):
    model.train()
    total_loss = 0

    for batch in loader:
        logits = model(
            batch["input_ids"].to(device),
            batch["attention_mask"].to(device),
            batch["audio"].to(device),
            batch["video"].to(device)
        )

        loss = loss_fn(logits, batch["label"].to(device))

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")
```

```python
model.eval()
y_true, y_pred = [], []

with torch.no_grad():
    for batch in loader:
        logits = model(
            batch["input_ids"].to(device),
            batch["attention_mask"].to(device),
            batch["audio"].to(device),
            batch["video"].to(device)
        )

        preds = torch.argmax(logits, dim=1).cpu().numpy()
        y_pred.extend(preds)
        y_true.extend(batch["label"].numpy())

acc = accuracy_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred, average="weighted")

print("🔥 Accuracy:", acc)
print("🔥 F1 Score:", f1)

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=labels, yticklabels=labels)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

```python
from transformers import AutoModel, AutoTokenizer

class TextEncoder:
    def __init__(self):
        self.tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")
        self.model = AutoModel.from_pretrained("bert-base-multilingual-cased")

    def encode(self, text):
        enc = self.tokenizer(text, return_tensors="pt", truncation=True, padding=True)
        out = self.model(**enc)
        return out.last_hidden_state  # sequence features
```

```python
from transformers import Wav2Vec2Model
import librosa
import torch

class AudioEncoder:
    def __init__(self):
        self.model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base-960h")

    def encode(self, path):
        try:
            y, sr = librosa.load(path, sr=16000)
            y = torch.tensor(y).unsqueeze(0)
            out = self.model(y).last_hidden_state
            return out
        except:
            return torch.zeros(1, 50, 768)
```

```python
import timm
import torch
import cv2
import numpy as np

class VideoEncoder:
    def __init__(self):
        self.model = timm.create_model("swin_tiny_patch4_window7_224", pretrained=True)
        self.model.head = torch.nn.Identity()

    def extract_frames(self, path):
        cap = cv2.VideoCapture(path)
        frames = []
        for _ in range(8):
            ret, frame = cap.read()
            if not ret:
                break
            frame = cv2.resize(frame, (224, 224))
            frames.append(frame)
        cap.release()

        if len(frames) == 0:
            return torch.zeros(8,3,224,224)

        frames = np.array(frames)
        frames = torch.tensor(frames).permute(0,3,1,2).float() / 255.0
        return frames

    def encode(self, path):
        frames = self.extract_frames(path)
        feats = self.model(frames)
        return feats.unsqueeze(0)  # (1, T, D)
```

```python
import torch.nn as nn
import torch.nn.functional as F
import torch

class MAG(nn.Module):
    def __init__(self, hidden_dim=768):
        super().__init__()
        self.W_hv = nn.Linear(hidden_dim*2, hidden_dim)
        self.W_ha = nn.Linear(hidden_dim*2, hidden_dim)
        self.W_v = nn.Linear(hidden_dim, hidden_dim)
        self.W_a = nn.Linear(hidden_dim, hidden_dim)
        self.beta = 1.0

    def forward(self, text, video, audio):

        # Align lengths (simple)
        min_len = min(text.size(1), video.size(1), audio.size(1))
        text = text[:, :min_len, :]
        video = video[:, :min_len, :]
        audio = audio[:, :min_len, :]

        weight_v = F.relu(self.W_hv(torch.cat((video, text), dim=-1)))
        weight_a = F.relu(self.W_ha(torch.cat((audio, text), dim=-1)))

        h_m = weight_v * self.W_v(video) + weight_a * self.W_a(audio)

        return text + h_m
```

```python
class TCLMAP_Model(nn.Module):
    def __init__(self, num_labels):
        super().__init__()

        self.text_encoder = TextEncoder()
        self.audio_encoder = AudioEncoder()
        self.video_encoder = VideoEncoder()

        self.mag = MAG()

        self.classifier = nn.Linear(768, num_labels)

    def forward(self, text, audio_path, video_path):

        text_feat = self.text_encoder.encode(text)
        audio_feat = self.audio_encoder.encode(audio_path)
        video_feat = self.video_encoder.encode(video_path)

        fused = self.mag(text_feat, video_feat, audio_feat)

        pooled = fused.mean(dim=1)
        return self.classifier(pooled)
```

```python
import torch
from torch.utils.data import Dataset
import pandas as pd
import librosa
import cv2
import numpy as np
from transformers import AutoTokenizer

class MultiModalDataset(Dataset):
    def __init__(self, excel_path, label_map):
        self.df = pd.read_excel(excel_path)
        self.tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")
        self.label_map = label_map

    def __len__(self):
        return len(self.df)

    def extract_audio(self, path):
        try:
            y, _ = librosa.load(path, sr=16000)
            y = y[:16000] if len(y) > 16000 else np.pad(y, (0, 16000 - len(y)))
            return torch.tensor(y, dtype=torch.float32)
        except:
            return torch.zeros(16000, dtype=torch.float32)

    def extract_video(self, path):
        try:
            cap = cv2.VideoCapture(path)
            frames = []

            for _ in range(5):  # fewer frames → faster
                ret, frame = cap.read()
                if not ret:
                    break
                frame = cv2.resize(frame, (64, 64))
                frames.append(frame)

            cap.release()

            if len(frames) == 0:
                raise Exception()

            frames = np.array(frames)

            if frames.shape[0] < 5:
                pad = np.zeros((5 - frames.shape[0], 64, 64, 3))
                frames = np.concatenate([frames, pad], axis=0)

            return torch.tensor(frames, dtype=torch.float32)
        except:
            return torch.zeros((5, 64, 64, 3), dtype=torch.float32)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        text = str(row["Hindi text"])
        label = self.label_map[str(row["intent"]).strip()]

        enc = self.tokenizer(
            text,
            padding="max_length",
            truncation=True,
            max_length=64,
            return_tensors="pt"
        )

        return {
            "input_ids": enc["input_ids"].squeeze(0),
            "attention_mask": enc["attention_mask"].squeeze(0),
            "audio": self.extract_audio(row["audio_path"]),
            "video": self.extract_video(row["video_path"]),
            "label": torch.tensor(label, dtype=torch.long)
        }
```
